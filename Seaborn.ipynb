```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,5)
```

```python

df = pd.read_csv("https://raw.githubusercontent.com/venuprakash17/CVR/refs/heads/main/Realistic_E-Commerce_Dataset.csv")
df.head()
```


## Derived Metrics (Using Pandas)

These metrics are **derived**, not assumed.


```python

df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.to_period("M").astype(str)

df["engagement_score"] = (
    df["pages_viewed"] +
    df["clicks"] +
    df["scroll_depth"] / 100
)

df["converted"] = df["purchase"].map({0: "No Purchase", 1: "Purchased"})
```

### ðŸ“Š Count Plot â€“ Category Comparison
- Use when comparing counts across categories
- Automatically handles categorical aggregation
- Best for frequency analysis

```python
sns.countplot(data=df, x="device_type")
plt.title("Device Type Distribution")
plt.show()
```
