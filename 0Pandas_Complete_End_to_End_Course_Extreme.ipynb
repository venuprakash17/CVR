## 3Ô∏è‚É£ Axis Explained (Most Confusing but Most Important)

### axis = 0 (DEFAULT)
- Column-wise operation
- Vertical direction

### axis = 1
- Row-wise operation
- Horizontal direction

If axis is misunderstood, aggregations will be wrong.


```python

num_cols = df.select_dtypes(include='number').columns

# Column-wise mean
df[num_cols].mean(axis=0)
```

```python

# Row-wise mean (per record)
df[num_cols].mean(axis=1).head()
```

```python

# Axis impact on sum
df[num_cols].sum(axis=0)
df[num_cols].sum(axis=1).head()
```

## 8Ô∏è‚É£ Handling Missing Values ‚Äì Numeric Data

### Mean Imputation
- Simple
- Sensitive to outliers

### Median Imputation
- Robust
- Preferred for skewed data


```python

df.fillna(df.mean(numeric_only=True)).head()
```

```python

df.fillna(df.median(numeric_only=True)).head()
```

## 9Ô∏è‚É£ Handling Missing Values ‚Äì Categorical Data

### Mode Imputation
Most frequent value replaces missing values.


```python

cat_cols = df.select_dtypes(include='object').columns

df_cat = df.copy()
df_cat[cat_cols] = df_cat[cat_cols].fillna(df_cat[cat_cols].mode().iloc[0])
df_cat.head()
```
