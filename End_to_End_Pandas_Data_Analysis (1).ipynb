## ðŸ”¹ 4. Data Cleaning & Preparation (CORE REAL-WORLD SKILL)

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/venuprakash17/CVR/refs/heads/main/Realistic_E-Commerce_Dataset.csv')
```

```python
df["age"] = df["age"].fillna(df["age"].median())
df["payment_method"] = df["payment_method"].fillna("Unknown")
```

```python
df.dropna(subset=["ad_campaign"], inplace=True)
```

```python
df.drop(columns=["ad_campaign"], inplace=True)
```

## ðŸ”¹ 6. Feature Engineering (EXPERT HABIT)

```python
df["date"].head()
```

```python
df["date"] = pd.to_datetime(df["date"])


df["day"] = df["date"].dt.day
df["month"] = df["date"].dt.month
df["weekday"] = df["date"].dt.day_name()
df["is_weekend"] = df["date"].dt.weekday >= 5
```

```python
df["hour"] = df["date"].dt.hour
df["minute"] = df["date"].dt.minute
df["second"] = df["date"].dt.second
```

```python
df["hour"]
```

```python
df["engagement_score"] = (
    df["pages_viewed"] +
    df["clicks"] +
    df["scroll_depth"]
)
```

```python
df.sort_values("age", ascending=False)
```

## ðŸ”¹ 8. Sorting, Ranking & Binning

# Why do we need Binning?

Binning is used to:

- Make data easier to understand

- Reduce noise in continuous data

- Enable group-level analysis

- Help in business decision-making

- Improve visualization & reporting

```python
# Binning (Segmentation)
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,18,30,45,60,100],
    labels=["Teen","Young","Adult","Mid","Senior"]
)
```

```python
df["time_of_day_bin"] = pd.cut(
    df["hour"],
    bins=[-1, 5, 11, 17, 23],
    labels=["Midnight", "Morning", "Afternoon", "Evening"]
)
```

## ðŸ”¹ 9. GroupBy & Aggregations (MOST IMPORTANT)

```python
df.groupby("product_category")["age"].mean()
```

```python
df.groupby("location").agg(
    avg_time_spent=("time_spent_minutes", "mean"),
    avg_pages=("pages_viewed", "mean"),
    total_purchases=("purchase", "sum")
)
```
